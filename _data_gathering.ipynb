```python
video_list = list(df['video_id'])
df_videos = pd.DataFrame(yt.get_video_metadata(video_list))
print(df_videos.shape)
df_videos.dropna(subset=['video_comment_count'], inplace=True)
print(df_videos.keys(), df_videos.shape)
video_list = list(df_videos['video_id'])
df_videos.head(31)
```

```python
cxv_list = []
for video_id in video_list:
    df_comments = pd.DataFrame(yt.get_video_comments(video_id, get_replies=False))
    cxv_list.append(df_comments)
    # print(len(cxv_list), df_comments.shape, video_id)
df_comments_all = pd.concat(cxv_list)
df_comments_all['publish_date'] = pd.to_timedelta(df_comments_all['comment_publish_date'], unit='s') + pd.to_datetime('1970-1-1')
df_comments_all
```
